```python
class Book:
  def __init__(self, title, author, isbn, price):
    self.title = title
    self.author = author
    self.isbn = isbn
    self.price = price

  def displayBookInfo(self):
    print(f"\nBook Title: {self.title},\nAuthor:{self.author},\nISBN No: {self.isbn},\nPrice: {self.price}")
```

```python
class Library:
  def __init__(self):
    self.books = []

  def addBook(self, book):
    self.books.append(book)
    print("Book added successfully...")

  def removeBook(self, book):
    self.books.remove(book)
    print("Book removed successfully...")

  def searchBook(self, title):
    for book in self.books:
      if book.title == title:
        return book
    return None

  def displayAvailableBooks(self):
    if len(self.books) == 0:
      print("No books available...")
      return
    print("\nAvailable Books...")
    for book in self.books:
      book.displayBookInfo()
```

```python
class User:
  def __init__(self, name, phone):
    self.name = name
    self.phone = phone
    self.borrowedBooks = []
    self.bill = 0

  def borrowBook(self, library, book):
    book = library.searchBook(book.title)
    if book is not None:
      self.borrowedBooks.append(book)
      self.bill += book.price
      library.removeBook(book)
      print("Book borrowed successfully...")
    else:
      print("Book not found...")

  def returnBook(self, library, book):
      self.borrowedBooks.remove(book)
      self.bill -= book.price
      library.addBook(book)
      print("Book returned successfully...")

  def displayBorrowedBooks(self):
    print("Borrowed Books...")
    for book in self.borrowedBooks:
      book.displayBookInfo()

  def totalBill(self):
    print(f"Total Bill: {self.bill}")

  def displayUserInfo(self):
    print("User Info...")
    print(f"\nName: {self.name},\nPhone: {self.phone}")
    self.displayBorrowedBooks()
    self.totalBill()
```

```python
def main():
  library = Library()

  while True:
    print('''\nWelcome to Library Management System
          \n
          1. Admin
          2. User
          3. Exit''')
    userType = int(input("Enter your choice: "))
    if userType == 1:
      while True:
        print('''\nMenu:
              1. Add Book
              2. Remove Book
              3. Search Book
              4. Display Available Books
              5. Exit''')
        choice = int(input("Enter your choice: "))
        if choice == 1:
          title = input("Enter book title: ")
          author = input("Enter book author: ")
          isbn = input("Enter book ISBN: ")
          price = float(input("Enter book price: "))
          book = Book(title, author, isbn, price)
          library.addBook(book)
        elif choice == 2:
          title = input("Enter book title: ")
          book = library.searchBook(title)
          if book is not None:
            library.removeBook(book)
          else:
            print("Book not found...")
        elif choice == 3:
          title = input("Enter book title: ")
          book = library.searchBook(title)
          if book is not None:
            book.displayBookInfo()
          else:
            print("Book not found...")
        elif choice == 4:
          library.displayAvailableBooks()
        else:
          break
    elif userType == 2:
      user = User(input("Enter user name: "), input("Enter user phone: "))
      while True:
        print('''\nMenu:
              1. Borrow Book
              2. Return Book
              3. Display User Info
              4. Exit''')
        choice = int(input("Enter your choice: "))

        if choice == 1:
          title = input("Enter book title: ")
          book = library.searchBook(title)
          if book is not None:
            user.borrowBook(library, book)
        elif choice == 2:
          title = input("Enter book title: ")
          book = library.searchBook(title)
          if book is not None:
            user.returnBook(library, book)
        elif choice == 3:
          user.displayUserInfo()
        elif choice >= 4:
          break
    else:
      print("Thank you for using Library Management System...")
      break
```

```python
if __name__=="__main__":
    main()
```
